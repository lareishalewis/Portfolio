1.List the names of top 10 actors who have acted in most films, include the number films.

```python
%%sql
SELECT actor_id,first_name,last_name,count(film_id) AS number_of_films
FROM actor JOIN film_actor USING(actor_id)
GROUP BY actor_id
ORDER BY number_of_films DESC
LIMIT 10
```

2.List the titles of all films on the inventory of Store 2 only, i.e. not found in Store 1.

```python
%%sql
SELECT film_id,title
FROM film
WHERE film_id IN(SELECT film_id
                    FROM inventory
                    WHERE store_id='2')
AND film_id NOT IN(SELECT film_id
                    FROM inventory
                    WHERE store_id='1')
```

3.List the names of all actors who have co-starred in the same movies with NATALIE HOPKINS.

```python
%%sql
WITH q AS
(SELECT actor_id,film_id
FROM actor JOIN film_actor USING(actor_id)
WHERE first_name = 'NATALIE' AND last_name='HOPKINS')

SELECT DISTINCT actor_id,first_name,last_name
FROM actor JOIN film_actor USING(actor_id)
WHERE actor_id NOT IN (SELECT actor_id FROM q)
AND film_id IN (SELECT film_id
                 FROM q)
```

6.List all orders placed by employees on their own birthdays. (HINT: You may use the MONTH and DAYOFMONTH functions)

```python
%%sql
SELECT *
FROM employee JOIN `order` USING(employee_id)
WHERE MONTH(order_date)=MONTH(employee_birthdate)
AND DAYOFMONTH(order_date)=DAYOFMONTH(employee_birthdate)
```

9.List the names of Products that are offered by only a single vendor, including the name of that vendor.

```python
%%sql
WITH q AS
(SELECT product_number,count(vendor_id) as c
FROM product_vendor
GROUP BY product_number
HAVING c=1)

SELECT product_number,product_name,vendor_id,vendor_name
FROM product NATURAL JOIN q
Natural JOIN product_vendor
NATURAL JOIN vendor
```

```python
%%sql

SELECT product.product_name,
                        product.product_number,
                        MIN(product_vendor.wholesale_price) AS min_price,
                        vendor.vendor_name
                  FROM shared_sales.product
                  INNER JOIN shared_sales.product_vendor
                  USING (product_number)
                  INNER JOIN shared_sales.vendor
                  USING (vendor_id)
                  GROUP BY product.product_number
                  ORDER BY product.product_number
```

```python
%%sql
WITH q AS
(SELECT product_number,MIN(wholesale_price) as p
FROM product_vendor
GROUP BY product_number)
SELECT product_number,product_name,vendor_id,vendor_name,wholesale_price
FROM product_vendor NATURAL JOIN q
NATURAL JOIN vendor
NATURAL JOIN product
WHERE wholesale_price=p
ORDER BY product_number
```

11.List all unique pairs of orders that have the same date, customer, and employee. (Hint: 10 pairs)

```python
%%sql
SELECT *
FROM `order` AS o1 JOIN `order` AS o2
ON o1.order_date = o2.order_date
AND o1.customer_id = o2.customer_id
AND o1.employee_id = o2.employee_id
WHERE o1.order_number < o2.order_number
```

12.Challenge: show all the triplets of products that appear together on at least 100 unique orders. (Hint: 5 triplets)

```python
%%sql

SELECT q1.product_number AS i1,q2.product_number AS i2,q3.product_number AS i3,count(*) AS n
FROM order__item as q1 JOIN order__item as q2 USING(order_number)
JOIN order__item as q3 USING(order_number)
WHERE q1.product_number < q2.product_number AND q2.product_number < q3.product_number
GROUP BY i1,i2,i3
HAVING n > 100
```

13.Challenge: Find the most common list of products on an order (Hint: The same list of 8 products is found on 23 orders. You may find the GROUP_CONCAT function helpful).

```python
%%sql
SELECT product_group, count(*) as count
FROM (SELECT order_number, GROUP_CONCAT(product_number ORDER BY product_number) as product_group
FROM order__item
GROUP BY order_number) as q
GROUP BY product_group
ORDER BY count DESC
LIMIT 1
```

# Hotel

14.Show all guests who have made reservations for any nights when Samantha Smith made reservations.

```python
%%sql
USE shared_hotel
```

```python
%%sql
SELECT guest_id,guest_name
FROM guest NATURAL JOIN reservation
WHERE guest_name != 'Samantha Smith'
AND date IN(SELECT date
FROM reservation
WHERE guest_id IN (SELECT guest_id FROM guest WHERE guest_name='Samantha Smith'))
GROUP BY guest_id
```

15.Challenge: Find the pair of guests who have made reservations for the same night the most. (Hint: two of the guests have made reservations for the same nights 20 times if you are using shared_hotels)

```python
%%sql
SELECT q1.guest_id as g1,q2.guest_id as g2,COUNT(*) as n
FROM reservation as q1 JOIN reservation as q2 USING(date)
WHERE q1.guest_id< q2.guest_id
GROUP BY g1,g2
ORDER BY n DESC
LIMIT 1
```
