```python
# Problem 1: List all stores including their full address and their manager names.
```

```python
%%sql

SELECT store.store_id, staff.first_name AS manager_first_name, staff.last_name AS manager_last_name, address.address, city.city, country.country, address.postal_code FROM store
INNER JOIN staff ON staff.staff_id = store.manager_staff_id
INNER JOIN address ON store.address_id = address.address_id
INNER JOIN city ON city.city_id = address.city_id
INNER JOIN country ON country.country_id = city.country_id;
```

```python
## Problem 2: List all payments over $11.00, including the payment date, the customer name, and staff name.
```

```python
%%sql

SELECT payment.payment_id, payment.payment_date, 
	   customer.first_name AS customer_first_name, 
       customer.last_name AS customer_last_name, 
       staff.first_name AS staff_first_name,
       staff.last_name AS staff_last_name,
       payment.amount FROM payment

INNER JOIN customer ON customer.customer_id = payment.customer_id
INNER JOIN staff ON staff.staff_id = payment.staff_id

WHERE payment.amount > 11.00 ;
```

```python
## Problem 3: List all customers who have made single payments of $11.00 or more. 
```

```python
%%sql

SELECT payment.payment_id, payment.payment_date, 
	   customer.first_name AS customer_first_name, 
       customer.last_name AS customer_last_name,
       payment.amount
       FROM payment

INNER JOIN customer ON customer.customer_id = payment.customer_id

WHERE payment.amount >= 11.00
```

```python
##problem 5: Display the product names, number ordered, and total amount owed for each item on order 160. 
```

```python
%%sql

SELECT order__item.order_number, product.product_name, order__item.quantity_ordered,
order__item.quoted_price AS quoted_price_each,
order__item.quantity_ordered * order__item.quoted_price AS total_amount_owed
	FROM product
INNER JOIN order__item ON order__item.product_number = product.product_number
WHERE order__item.order_number = 160;
```

```python
## Problem 16: Show the names of guest who have made reservations but have not checked in. 
```

```python
%%sql 

SELECT guest_name FROM guest WHERE guest_id IN(
SELECT guest_id FROM reservation WHERE (room, `date`) NOT IN(
SELECT DISTINCT room, `date` FROM check_in)
ORDER BY `date` ASC, room ASC)
```

```python
## Problem 17: Show the names of guest who have checked into Deluxe rooms and have not yet checked out.  
```

```python
%%sql

SELECT guest_name FROM guest WHERE guest_id IN(
    SELECT guest_id FROM reservation WHERE (room, `date`) IN(
        SELECT DISTINCT check_in.room, `date` FROM check_in
        INNER JOIN room ON room.room=check_in.room
        WHERE (check_in.room, `date`) NOT IN(
            SELECT DISTINCT check_out.room, check_out.date 
            FROM check_out
            INNER JOIN room ON room.room=check_out.room
            WHERE room_type = 'Deluxe')
        AND room_type = 'Deluxe'))
```

```python
## Problem 18: Show the names of the 10 guests who checked in most recently. 
```

```python
%%sql

SELECT guest.guest_name, checkins.`date`, checkins.room FROM guest 
INNER JOIN(
    SELECT DISTINCT room, guest_id, date FROM reservation WHERE (room, `date`) IN(
        SELECT * FROM check_in 
        ORDER BY `date` DESC)
    ORDER BY reservation.`date` DESC, reservation.room ASC LIMIT 10)
    AS checkins ON checkins.guest_id = guest.guest_id
```
