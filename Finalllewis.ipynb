```python
@schema
class CarAvailable(dj.Manual):
    definition = """
    -> Car
    car_available_date : date
    ---
    car_price : decimal(6, 2)
    free_miles_per_day : int
    extra_price_per_mile : decimal(6,2)
    """
```

```python
start_date = datetime.date(2021, 12, 1)
days = 61


for day in tqdm.tqdm(range(days)):
    car_price = random.randint(40, 350)
    free_miles=random.randint(45,60)
    extra_per_mile=random.randint(20, 40)
    CarAvailable.insert(dict(key, 
             car_available_date=start_date + datetime.timedelta(days=day), 
             car_price=car_price,
             free_miles_per_day=free_miles,
             extra_price_per_mile=extra_per_mile) for key in Car.fetch('KEY'))
    
```
